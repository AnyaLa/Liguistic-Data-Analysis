#### NE count (target variable)

```python
def get_ne(text):
    #ne = re.findall(r'&[\w ]+!&', text)
    ne = re.findall(r'&(.*?)!&', text)
    return ne
```

```python
ne0 = get_ne(marked_texts.text.values[10])
```

```python
genre_map = {'FAQ':'Статьи', 
             'talks':'Лекции', 
             'видео':'Лекции', 
             'журнал':'Статьи'}
```

```python
def genre_mapping(list):
    genre_list = []
    for genre in list:
        try:
            genre = genre_map[genre]
        except KeyError:
            pass
        genre_list.append(genre)
    return genre_list
```

```python
genre_l = genre_mapping(r_ner_data.Genre.values)
```

```python
r_ner_data.Genre = genre_l
```
